#Social Network Ads

In this exercise, we analyse data from a given dataset containing a userid, gender, salary and purchased label. KNN will be used as our classifier.

```python
import pandas as pd
import numpy as np
```

```python

df=pd.read_csv('https://raw.githubusercontent.com/JuliaKav/SocialNetworkAds/main/7.1%20Social_Network_Ads.csv.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
X=df.iloc[:,[2,3]].values
```

```python
X
```

```python
y=df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

```

```python
kn = KNeighborsClassifier(n_neighbors=5,metric='minkowski', p=2)
kn.fit(X_train,y_train)
pred=kn.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```
